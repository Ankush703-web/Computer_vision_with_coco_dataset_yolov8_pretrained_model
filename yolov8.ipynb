```python
from ultralytics import YOLO
```

```python
# Load a pretrained YOLOv8 model
model = YOLO('yolov8n.pt')
```

```python
# Define the source image path or URL
image_path = 'https://ultralytics.com/images/zidane.jpg'

# Run object detection on the image
results = model.predict(image_path)

# Print the results
print(results)
```

```python
# Display the image with bounding boxes (optional)
from PIL import Image
import matplotlib.pyplot as plt

# Open the image using PIL
image = Image.open('/content/images.jpg')

# Get the annotated image from the results
annotated_image = results[0].plot()

# Display the annotated image using matplotlib
plt.imshow(annotated_image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
import os

# Load the YOLOv8 model
model = YOLO('yolov8n.pt')

# Path to the video file
video_path = '/content/drive/MyDrive/VID-20250411-WA0227.mp4'

# Open the video file
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define output path and VideoWriter
output_path = '/content/output.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Process the video frame-by-frame
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run YOLOv8 inference on the frame
    results = model(frame)

    # Annotate the frame
    annotated_frame = results[0].plot(labels=False)

    # Write the annotated frame to output video
    out.write(annotated_frame)

# Release everything
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
video_capture = cv2.VideoCapture('/content/output.mp4')

if not video_capture.isOpened():
    raise IOError("Cannot open webcam")

ret, frame = video_capture.read()
cv2_imshow(frame)

if cv2.waitKey(1) & 0xFF == ord('q'):
  break

video_capture.release()
cv2.destroyAllWindows()
```
